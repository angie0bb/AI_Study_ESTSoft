#### Stationary
- 아래 그래프를 보면 시간의 흐름에 따라 상향하는 추세가 보임 -> non-stationary 의심
- rolling mean, std도 constant하진 않다.
- 데이터를 보면 1월 1일에 매출이 거의 없음. 

```python
# Stationary check
plt.plot(df, label = "sales")
plt.plot(df.rolling(window=30).mean(), label = "rolling_avg_sales")
plt.plot(df.rolling(window=30).std(), label = "rolling_std_sales") 
plt.show() # 상향하는 추세가 실찍 보임, non-stationary 의심. rolling mean, std도 constant하진 않다.
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig = plt.figure(figsize=(12, 8)) 
ax1 = fig.add_subplot(211)
plot_acf(df, lags = 50, ax = ax1)
ax2 = fig.add_subplot(212)
plot_pacf(df, lags = 50, ax = ax2)
plt.show()
```

```python
# adf test 
from statsmodels.tsa.stattools import adfuller

adf = adfuller(df)  
print('ADF Statistic: {}'.format(adf[0]))
print('p-value: {}'.format(adf[1]))
print('Critical Values:')
for key, value in adf[4].items():
  print('\t{}: {}'.format(key, value))
```

p-value = 0.09 이니까 사실 CI 90% 기준으로는 stationary하다고 볼 수 있음.

```python
# train, test 분리

trn, tst = df[:1461], df[1461:] # 3년치 자름 -> 근데 정확히 4년은 아니고 3년 반정도니까..흠?
trn
```

```python
# 1st Model: ARIMA - pacf를 보면 7일 기점에서 확 꺾임. (아마 주간 - 평일/주말 영향이 있지 않을까)
from statsmodels.tsa.arima.model import ARIMA
model_1 = ARIMA(trn, order = (7,1,0)).fit()
```

```python
pred = model_1.predict("2017-01-01", "2017-08-15")
tst.plot()
pred.plot(label="Prediction")
plt.legend()
plt.show()
```

```python
# auto arima
from pmdarima.arima import auto_arima

model_auto = auto_arima(trn, start_p=0, start_q=0, max_p=15, max_q=15, trace=True)
```

```python
pred_auto = model_auto.predict(200)  # Best model:  ARIMA(12,1,4)(0,0,0)[0] intercept
tst.plot()
pred_auto.plot(label="Prediction")
plt.legend()
plt.show()
```
